```python
# ============================================================
# ============================================================
# Transfer Learning
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
# How to use a model that has been trained for one task for another.
# By using a model that has been pre-learned on large scale data, 
# even a small amount of training data can be used to obtain sufficient performance.
# ============================================================
# ============================================================
```
